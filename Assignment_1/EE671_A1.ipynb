```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("Q1a.txt")

time = data[:, 0]
vin = data[:, 1]
vout = data[:, 3]

time_ns = time * 1e9

plt.figure(figsize=(10, 6))
plt.plot(time_ns, vin, label="Vin (V)", linewidth=2)
plt.plot(time_ns, vout, label="Vout (V)", linewidth=2)
plt.xlabel("Time (ns)")
plt.ylabel("Voltage (V)")
plt.title("INVX1 loaded with INVX1 Input and Output Waveforms")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the VTC data (Vin, Vout)
data = np.loadtxt("Q1b.txt")
vin, vout = data[:, 0], data[:, 1]

# Compute derivative dVout/dVin
dvout_dvin = np.gradient(vout, vin)

# Switching threshold Vm: where Vout = Vin
idx_vm = np.argmin(np.abs(vout - vin))
Vm = vin[idx_vm]

# Find V_IL, V_IH (where slope ~ -1)
idx_cross = np.where(np.isclose(dvout_dvin, -1, atol=0.05))[0]
Vil, Vih = None, None
if len(idx_cross) > 0:
    Vil = vin[idx_cross[0]]
    Vih = vin[idx_cross[-1]]

# Noise margins
if Vil and Vih:
    Nml = Vil
    Nmh = 1.8 - Vih
else:
    Nml, Nmh = None, None

# ---------------- Plot VTC ----------------
plt.figure(figsize=(7,6))
plt.plot(vin, vout, label="Vout vs Vin", linewidth=2)
plt.plot(vin, vin, 'r--', label="Unity Slope (Vout=Vin)") # slope=1 line

# Mark Vm
plt.scatter(Vm, vout[idx_vm], color="black", zorder=5, label=f"Vm = {Vm:.2f} V")

# Mark Vil, Vih if found
if Vil: plt.scatter(Vil, np.interp(Vil, vin, vout), color="green", label=f"Vil = {Vil:.2f} V")
if Vih: plt.scatter(Vih, np.interp(Vih, vin, vout), color="blue", label=f"Vih = {Vih:.2f} V")

plt.xlabel("Vin (V)")
plt.ylabel("Vout (V)")
plt.title("INVX1 VTC")
plt.grid(True)
plt.legend()
plt.show()

# Print results
print(f"Vm = {Vm:.3f} V")
if Vil: print(f"Vil = {Vil:.3f} V")
if Vih: print(f"Vih = {Vih:.3f} V")
if Nml and Nmh:
    print(f"Nml = {Nml:.3f} V")
    print(f"Nmh = {Nmh:.3f} V")
```
